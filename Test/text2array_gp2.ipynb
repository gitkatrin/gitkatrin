```python
from transformers import GPT2Tokenizer, GPT2Model
import torch
import numpy as np

# Load pre-trained model tokenizer (vocabulary)
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
# Load pre-trained model (weights)
model = GPT2Model.from_pretrained('gpt2')

# Encode the input text
text = "No driving on shoulders."
inputs = tokenizer(text, return_tensors="pt")

# Generate the word embedding vectors
outputs = model(**inputs)
text_array = outputs.last_hidden_state

# Convert last_hidden_states to a NumPy array
text_array_np = text_array.detach().numpy()

# Save the NumPy array to a file
np.save('text_array.npy', text_array_np)

# Print the shape of the word embedding vectors
print(text_array.shape)
print(text_array_np)


# Access individual word embeddings
# For example, the embedding for the first token (word) in the sentence:
first_word_embedding = text_array[0][0]
first_word_embedding

```
